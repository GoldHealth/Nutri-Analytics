```python
# Primer año
primero = vimc[vimc.curso == 1]
imc1 = primero.IMC

c1 = 0
c2 = 0
c3 = 0
c4 = 0
for i in imc1:
    if i < 18.50:
        c1 = c1 + 1
    elif i >= 18.50 and i <= 24.90:
        c2 = c2 + 1
    elif i >= 25.00 and i <= 29.90:
        c3 = c3 + 1
    else:
        c4 = c4 + 1
# Porcentajes

t1 = c1 + c2 + c3 + c4

B1 = c1 / t1*100
N1 = c2 / t1*100
S1 = c3 / t1*100
O1 = c4 / t1*100

I1 = [B1, N1, S1, O1]
```

```python
# segundo año
segundo = vimc[vimc.curso == 2]
imc2 = segundo.IMC

A1 = 0
A2 = 0
A3 = 0
A4 = 0
for i in imc2:
    if i < 18.50:
        A1 = A1 + 1
    elif i >= 18.50 and i <= 24.90:
        A2 = A2 + 1
    elif i >= 25.00 and i <= 29.90:
        A3 = A3 + 1
    else:
        A4 = A4 + 1
# Porcentajes

t2 = A1 + A2 + A3 + A4

B2 = A1 / t2*100
N2 = A2 / t2*100
S2 = A3 / t2*100
O2 = A4 / t2*100

I2 = [B2, N2, S2, O2]
```

```python
# tercero año
tercero = vimc[vimc.curso == 3]
imc3 = tercero.IMC

b1 = 0
b2 = 0
b3 = 0
b4 = 0
for i in imc3:
    if i < 18.50:
        b1 = b1 + 1
    elif i >= 18.50 and i <= 24.90:
        b2 = b2 + 1
    elif i >= 25.00 and i <= 29.90:
        b3 = b3 + 1
    else:
        b4 = b4 + 1
# Porcentajes

t3 = b1 + b2 + b3 + b4

B3 = b1 / t3*100
N3 = b2 / t3*100
S3 = b3 / t3*100
O3 = b4 / t3*100

I3 = [B3, N3, S3, O3]
```

```python
# https://www.delftstack.com/es/howto/matplotlib/stack-bar-plots-matplotlib/
state = {
"Bajo" : [B, B1, B2, B3],
"Normal" : [N, N1, N2, N3],
"Sobrepeso" : [S, S1, S2, S3],
"Obesidad" : [O, O1, O2, O3],

}
year=["General","Primero","Segundo", "Tercero"]

df=pd.DataFrame(state,index=year)


df.plot(kind="bar", title = "Estado IMC por Clase", xlabel = "Poblacion", ylabel = "Porcentaje", stacked=True, figsize=(10,9))

plt.legend(loc="lower left",bbox_to_anchor=(0.6,0.8))
plt.show()
```
